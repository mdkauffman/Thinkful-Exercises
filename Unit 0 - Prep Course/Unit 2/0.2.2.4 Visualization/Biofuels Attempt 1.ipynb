```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('totalbiofuelsconsumption.csv')
df.set_index("Country", inplace = True) 
biofuels = df
biofuels
```
