```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

norm = np.random.normal(0, 1, 100)
norm.sort()
```

```python
a = np.random.geometric(.1, 100)
plt.hist(a)
# Add a vertical line at the mean.
plt.axvline(a.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one standard deviation above the mean.
plt.axvline(a.mean() + a.std(), color='b', linestyle='dashed', linewidth=2)
# Add a vertical line at one standard deviation below the mean.
plt.axvline(a.mean()-a.std(), color='b', linestyle='dashed', linewidth=2) 
# Print the histogram.
plt.show()

a.sort()
plt.plot(norm, a, "o") 
plt.show() 
```
