```python
# removing the £ and $ from the cost column
remove_poundsign = wellcome['Cost'].apply(lambda x: str(x).replace('£',''))

wellcome['Cost'] = remove_poundsign

remove_dollarsign = wellcome['Cost'].apply(lambda x: str(x).replace('$',''))

wellcome['Cost'] = remove_dollarsign
```

```python
# converting the cost to numeric values
wellcome.Cost = pd.to_numeric(wellcome.Cost, errors='coerce')
```

```python
# Checking to see what outliers are present, both above and below the range.
# wellcome[wellcome['Cost']>4000]
# wellcome[wellcome['Cost']<100]
```

```python
# removing lines with costs of 999,999
remove_cost9 = wellcome[wellcome.Cost != [999999]]
wellcome_costcalc = remove_cost9


# cleaning some of the higher (>10,000) cost values by multiplying them by .01; 
# the assumption is that they were mistakenly punched in without differentiating between pounds and pence.
wellcome_costs = wellcome_costcalc['Cost']
mod_costs = []
for x in wellcome_costs:
    if x >10000:
        mod_costs.append(x * .01)
    else:
        mod_costs.append(x)
wellcome_costs = mod_costs
wellcome_costcalc['Cost'] = wellcome_costs
```

```python
# Making a new dataframe containing the list of individual journal titles

journal_list = []

for x in wellcome_costcalc['Journal title']:
    if x in journal_list:
        continue
    journal_list.append(x)

journal_list_df = pd.DataFrame({'Journal_title':journal_list})
# print (journal_list_df)
```

```python
# calculating the mean cost of articles for each journal, and adding that list to the journal dataframe
mean_costs = []

for x in journal_list:
    journal_costs = wellcome_costcalc[wellcome_costcalc['Journal title'] == x]
    mean_costs.append(journal_costs['Cost'].mean())

journal_list_df['Mean_article_cost'] = mean_costs

```

```python
# calculating the median cost of articles for each journal, and adding that list to the journal dataframe
median_costs = []

for x in journal_list:
    journal_costs = wellcome_costcalc[wellcome_costcalc['Journal title'] == x]
    median_costs.append(journal_costs['Cost'].median())

journal_list_df['Median_article_cost'] = median_costs
```

```python
# calculating the standard deviation of costs of articles for each journal, and adding that list to the journal dataframe
cost_stdev = []

for x in journal_list:
    journal_costs = wellcome_costcalc[wellcome_costcalc['Journal title'] == x]
    cost_stdev.append(journal_costs['Cost'].std())

journal_list_df['Article_cost_stdev'] = cost_stdev

journal_list_df.head(1000)
```
