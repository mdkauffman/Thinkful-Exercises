This distribution was chosen to create a singular blob with three barely distinguishable lobes, and find out how each algorithm handles that situation.

#### Version 1, K Means

That's acutally remarkably good.  Mind you, it only knows to look for three lobes because I told it so, and thus it can guess where the centers of the blobs are, but even given that it still performs a lot better than I was expecting, given it's utterly moronic performance with the heart disease data.

#### Version 1, Mean Shift

This isn't very impressive.  I guess it could detect that there was a wider blob on the bottom, and a narrow blob on the top, and just drew a line between them.  Without knowing to look for three centers, it just guessed wrong.  Good to know, that Mean Shift doesn't deal well with multiple merged blobs.

#### Version 1, Spectral Clustering

Yeah, so if Mean Shift doesn't deal _well_ with merged blobs, then Spectral Clustering doesn't deal with it at _all_.  It just assumes that the entire mess is a single blob, and guesses that four of the outliers are separate blobs!  Definitely worth keeping an eye on this one, in real-life trials, to make sure it doesn't over-generalize.

#### Version 1, Affinity Propagation

Yeah, this wasn't any kind of surprise, given how it performed in the lesson demo.  The cluster sizes it goes for are way too small to make sense of blobs with as many points as these have; is there a way to increase the number of connected points that the affinity algorithm clusters together?

Dang.  Provided you have an accurate count of the number of blobs there "should" be, and they're all reasonably centered on their own point, then K-Means seems like a really reliable algorithm.

#### Version 3, Mean Shift

And this is why I was expecting this setup to be challenging; Mean Shift just doesn't deal with blurry boundaries well.  Even the two on the bottom left get lumped together, which is really astonishing to me, I thought they'd be gimmes for this algorithm.

I was half-expecting the top right blobs to get treated like Spectral did to the Version 1 setup, all grouped together with a few outlier points selected as the 9th and 10th blobs.  But no, Spectral actually did a really good job on this setup, barring a few missed points that cross the boundary between two merged blobs.

#### Version 3, Affinity Propagation

Well now, we're making progress!  It's only twice as many blobs as are actually present, rather than four or twelve.  It's actually being somewhat accurate, too; the upper-right-most cyan blob is seemingly almost entirely correct, though the green one to it's left is stealing a few points from it.

## Version 1 (3x; 10-point spread; 7 std)

```python
# The coordinates of the centers of our blobs.
centers = [[0, 10], [-10, -10], [10, -10]]

# Make 10,000 rows worth of data with two features representing three
# clusters, each having a standard deviation of 1.
X, y = make_blobs(
    n_samples=10000,
    centers=centers,
    cluster_std=7,
    n_features=2,
    random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()

#Divide into training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.9,
    random_state=42)
```

#### Version 1, K Means

```python
y_pred = KMeans(n_clusters=3, random_state=42).fit_predict(X_train)

# Plot the solution.
#plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred)
#plt.show()

# Check the solution against the data.
# Plot the solution.
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_pred)
plt.show()

# Check the solution against the data.
print('Comparing k-means clusters against the data:')
print(pd.crosstab(y_pred, y_train))
```

#### Version 1, Mean Shift

```python
# Here we set the bandwidth. This function automatically derives a bandwidth
# number based on an inspection of the distances among points in the data.
bandwidth = estimate_bandwidth(X_train, quantile=0.2, n_samples=500)

# Declare and fit the model.
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X_train)

# Extract cluster assignments for each data point.
labels = ms.labels_

# Coordinates of the cluster centers.
cluster_centers = ms.cluster_centers_

# Count our clusters.
n_clusters_ = len(np.unique(labels))

print("Number of estimated clusters: {}".format(n_clusters_))

plt.scatter(X_train[:, 0], X_train[:, 1], c=labels)
plt.show()

print('Comparing the assigned categories to the ones in the data:')
print(pd.crosstab(y_train,labels))
```

#### Version 1, Spectral Clustering

```python
# We know we're looking for three clusters.
n_clusters=3

# Declare and fit the model.
sc = SpectralClustering(n_clusters=n_clusters)
sc.fit(X_train)

#Predicted clusters.
predict=sc.fit_predict(X_train)

#Graph results.
plt.scatter(X_train[:, 0], X_train[:, 1], c=predict)
plt.show()

print('Comparing the assigned categories to the ones in the data:')
print(pd.crosstab(y_train,predict))
```

#### Version 1, Affinity Propagation

```python
# Declare the model and fit it in one statement.
# Note that you can provide arguments to the model, but we didn't.
af = AffinityPropagation().fit(X_train)

# Pull the number of clusters and cluster assignments for each data point.
cluster_centers_indices = af.cluster_centers_indices_
n_clusters_ = len(cluster_centers_indices)
labels = af.labels_

print('Estimated number of clusters: {}'.format(n_clusters_))

plt.figure(1)
plt.clf()

# Cycle through each cluster and graph them with a center point for the
# exemplar and lines from the exemplar to each data point in the cluster.
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = X_train[cluster_centers_indices[k]]
    plt.plot(X_train[class_members, 0], X_train[class_members, 1], col + '.')
    plt.plot(cluster_center[0],
             cluster_center[1],
             'o',
             markerfacecolor=col,
             markeredgecolor='k')
    for x in X_train[class_members]:
        plt.plot([cluster_center[0], x[0]], [cluster_center[1], x[1]], col)

plt.title('Estimated number of clusters: {}'.format(n_clusters_))
plt.show()

os.system('say "all done."'); print('\a')  # this could take a while, let me know when it's done
```
